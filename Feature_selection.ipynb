**There are 3 Best techniques for feature selection**

- `constant technique`
- `quasi constant technique `
- `Hypothesis Testing `

**constant**

```python
# constant = we are going to remove the columns which are having variance zero
```

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
reg = VarianceThreshold(threshold = 0.0)
```

```python
df = df.iloc[: , :-1]
t = df.iloc[: , -1]
df
```

```python
reg.fit(df)
```

```python
sum(reg.get_support())
```

```python
df.columns[reg.get_support()]
```

```python
df.columns[~reg.get_support()]
```

```python
df['var_23'].var()
```

```python
s = df.columns[reg.get_support()]
s
```

```python
df = reg.transform(df)
df
```

```python
df = pd.DataFrame(df,columns=s)
df.shape
```

**Hypothesis Testing**

```python
from scipy.stats import pearsonr
```

```python
df.columns
```

```python
c = []
for i in df.columns:
  values = pearsonr(df[i],t)
  c.append(values)

c
```

```python
c = np.array(c)
c
```

```python
p_value = pd.Series(c[:,1] , index=df.columns)
p_value
```

```python
p_value.sort_values(ascending=False , inplace=True)
```

```python
p_value.plot.bar()
```
