```python
df = pd.read_csv('/content/Titanic-Dataset.csv',usecols = ['Age','Fare'])
df
```

```python
def fun():
  plt.figure(figsize=(8,3))
  plt.subplot(1,2,1)
  plt.title('Normal Distribution')
  df['Fare'].plot(kind='kde',color='r')

  plt.subplot(1,2,2)
  plt.title('Outliers')
  sns.boxplot(x = df['Fare'])

  plt.show()
```

```python
fun()
```

**varibale Transformations**

- `If the Numerical column is following the Normal Distribution then (there will be less amount of outliers ) -> and we can say that Numerical column is a good column `

- `If the Numerical column is not following the Normal Distribution (then we have more outliers ) -> which is not good to train the Algorithm -> In that case we need to use (Varibale Transformation techniques) -> and Make the Numerical column as the best column `

ex: [Age] -> [ND] -> [Less outliers] -> Important column

    [Fare] -> [not following the ND] -> [More outliers] -> [using varibale transformation technique] -> [make fare column to follow] -> [ND]

**varibale Transformation Techniques**

- `Log Transformation `
- `reciprocal Transformation`
- `exponential Transformation`
- `square root Transformation`
- `box cox Transformation`
- `yeojohnson Transformation`

```python
df = df.dropna()
```

```python
df
```

```python
from scipy import stats
```

```python
def fun(df,var):
  plt.figure(figsize=(11,3))

  plt.subplot(1,3,1)
  plt.title('Normal Distribution')
  df[var].plot(kind='kde',color='r')

  plt.subplot(1,3,2)
  plt.title('Outliers')
  sns.boxplot(x = df[var])

  plt.subplot(1,3,3)
  stats.probplot(df[var],dist='norm',plot=plt)

  plt.show()
```

```python
fun(df,'Fare')
```

```python
# Log
import warnings
warnings.filterwarnings('ignore')
df['Fare_log'] = np.log(df['Fare'])
```

```python
fun(df,'Fare_log')
```

```python
# reciprocal
df['Fare_reci'] = 1 / df['Fare']
fun(df,'Fare_reci')
```

```python
# exponential
df['Fare_exp'] = np.exp(df['Fare'])
fun(df,'Fare_exp')
```

```python
# square root
df['Fare_sqrt'] = np.sqrt(df['Fare'])
fun(df,'Fare_sqrt')
```

```python
# boxcox
df['Fare_boxcox'],alpha = stats.boxcox(df['Fare'])
```

```python
alpha
```

```python
fun(df,'Age_boxcox')
```

```python
# Yeojohnson

df['Age_yeo'],alpha = stats.yeojohnson(df['Age'])
```

```python
fun(df,'Age_yeo')
```
