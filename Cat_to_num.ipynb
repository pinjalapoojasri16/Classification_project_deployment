```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'Gender':['Male','Female','Male','Female'],
                   'Region':['East','West','East','North'],
                   'Rented_own_house':['Rent','Own','Rent','Own'],
                   'Occupation':['SE','Developer','Testing','AI'],
                   'Education':['PHD','BTECH','Degree','MSC']
                   })
df
```

```python
# For nominal columns = nominal encoding or OnehotEncoding technique
# For Odinal columns = Odinal encoding
# for dependent column = Label Encoding
```

```python
from sklearn.preprocessing import OneHotEncoder
one_hot = OneHotEncoder(drop='first')
one_hot.fit(df[['Gender','Region']])
```

```python
one_hot.categories_
```

```python
one_hot.get_feature_names_out()
```

```python
result = one_hot.transform(df[['Gender','Region']]).toarray()
result
```

```python
f = pd.DataFrame(data = result,columns=one_hot.get_feature_names_out(),
                 index=['r1','r2',2,'r3'])

f
```

```python
df.reset_index(drop=True, inplace=True)
f.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
f
```

```python
final_result = pd.concat([df,f],axis=1)
final_result = final_result.drop(['Gender','Region'],axis=1)
```

```python
from sklearn.preprocessing import OrdinalEncoder
od = OrdinalEncoder()
od.fit(df[['Rented_own_house', 'Occupation', 'Education']])
```

```python
od.categories_
```

```python
c_names = od.get_feature_names_out()
c_names = c_names + '_Od'
c_names
```

```python
r = od.transform(df[['Rented_own_house', 'Occupation', 'Education']])
r
```

```python
g = pd.DataFrame(data = r,columns = c_names)
g
```

```python
final_result = pd.concat([final_result,g],axis=1)
final_result = final_result.drop(['Rented_own_house', 'Occupation', 'Education'],axis=1)
final_result
```
